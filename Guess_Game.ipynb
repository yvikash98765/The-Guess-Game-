```python
import random 
# Library that we use in order to choose
# one random words from list of words

name = input("What is your Name")
# In this the user is asking the name of the player

print("Welcome to the game",name)
print("Get ready for Game will start soon")

words = ['programming', 'computer', 'python', 'mathematics', 'player', 'rainbow', 'reverse', 'forward', 'condition', 'geeks', 'board', 'water']

word = random.choice(words)
# This function will use one random word from the list of words

print("Guess the Characters of word")

guesses = ''

#any number of turns can be used here
turns = 12

while turns > 0:
    
  failed = 0 	
	# counts the number of times a user fails 
	
	# all characters from the input word taking one at a time. 
	
  for char in word:

    # comparing that character with 
		# the character in guesses 
    
    if char in guesses:
		   print(char) 	
    
    else: 
       print("_") 
			 
       failed = failed + 1
      # for every failure 1 will be 
			# incremented in failure

  if failed == 0: 
		# user will win the game if failure is 0 and 'You Win' will be given as output 
		
    print("You Win") 

    print("The word is: ", word) 
		# this print the correct word 
    break
	
	# if user has input the wrong alphabet then it will ask user to enter another alphabet 
  guess = input("guess a character:") 
	
  guesses = guesses + guess
  # every input character will be stored in guesses  
	
	# check input with the character in word 
  if guess not in word: 

	  turns = turns - 1
		
		# if the character doesn’t match the word then “Wrong” will be given as output 
	  print("Wrong") 
   
  print('You have ', + turns, 'more guesses')
		# this will print the number of turns left for the user

  if turns == 0: 
			print("You Loose") 
```
